```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# step 1: Load the dataset
df = pd.read_csv('tips.csv')


# step 2: One-hot encode categorical variables
df_encoded = pd.get_dummies(df, columns=['sex', 'smoker', 'day', 'time'], drop_first=True)
df_encoded = df_encoded.astype(int)
print(df_encoded)

# step 3: Define features and target
X = df_encoded.drop('tip', axis=1)
y = df_encoded['tip']
print(y)

#  step 4 Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#  step 5:Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# step 6: Build ANN model
model = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),
    Dense(32, activation='relu'),
    Dense(1)
])

# step 7: Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# step 8: Train model
model.fit(X_train, y_train, epochs=100, batch_size=10, verbose=1, validation_split=0.1)

#step 9:  Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test Accuracy:", accuracy)

```

```python

# Your test data
customers = np.array([[ 24,3,4,0,0,0,1,0,0]])

customers = scaler.transform(Customers)

# Predict using the trained model
pred = model.predict(customers).flatten()
print(f"Predicted tip for custom input: ${pred[0]:.2f}")
```
